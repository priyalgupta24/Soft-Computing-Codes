```
trainingDataSet=data.iloc[0:trainingDataLen,:]
trainingDataSet
```

```
testingDataSet=data.iloc[trainingDataLen:,:]
```

```
def removeDuplicate(txtStr):
  ans=[]
  for num in txtStr:
    if num not in ans:
      ans.append(num)
  return ans

# duplicate = [2, 4, 10, 20, 5, 2, 20, 4] 
# print(removeDuplicate(duplicate))
```

```
def frequency(trainingDataSet):
  spam={}
  ham={}
  dt=trainingDataSet.iloc[:,1]
  category=trainingDataSet.iloc[:,0]
  for i in range(0,trainingDataLen):
    if(category.iloc[i]=='spam'):
      txtStr = dt.iloc[i].split(" ")
      txtStr = removeDuplicate(txtStr)
      for j in txtStr:
        if(spam.get(j)==None):
          spam[j]=1
        else:
          spam[j]+=1
    elif(category.iloc[i]=='ham'):
      txtStr = dt.iloc[i].split(" ")
      txtStr = removeDuplicate(txtStr)
      for j in txtStr:
        if(ham.get(j)==None):
          ham[j]=1
        else:
          ham[j]+=1
  
  for key, value in spam.items():
    if(ham.get(key)==None):
      ham[key]=0
  for key, value in ham.items():
    if(spam.get(key)==None):
      spam[key]=0

  return spam, ham
    
```

```
spam, ham = frequency(trainingDataSet)
ham
```

```
def probability(spam,ham):
  spamProb={}
  hamProb={}
  tot=len(spam)
  for key, value in spam.items():
    # print(key," ", value)
    spamProb[key]=value/tot
  tot=len(ham)
  for key, value in ham.items():
    hamProb[key]=value/tot
  return spamProb, hamProb
```

```
spamProb, hamProb = probability(spam,ham)
spamProb
```

```
spamLen=len(spam)
numSpam=0
category=trainingDataSet.iloc[:,0]
for i in range(0,trainingDataLen):
    if(category.iloc[i]=='spam'):
      numSpam+=1
pSpam=numSpam/trainingDataLen
pHam=1-pSpam
print(pSpam,pHam)
```

```

def testDataAccuracy(spamProb, hamProb, testingDataSet):
  ans={}
  count=0
  actualSpam=0
  preSpam=0
  actualHam=0
  preHam=0
  dt=testingDataSet.iloc[:,1]
  category=testingDataSet.iloc[:,0]
  testDataLen=len(testingDataSet)
  for i in range(0,testDataLen):
    stringGivenSpam=1
    stringGivenHam=1
    str1=dt.iloc[i]
    txtStr = dt.iloc[i].split(" ")
    txtStr = removeDuplicate(txtStr)
    for key, value in spamProb.items():
      if key in txtStr:
        stringGivenSpam=stringGivenSpam*value
      else:
        stringGivenSpam=stringGivenSpam*(1-value)
    

      # txtStr = dt.iloc[i].split(" ")
      # txtStr = removeDuplicate(txtStr)
    for key, value in hamProb.items():
      if key in txtStr:
        stringGivenHam=stringGivenHam*value
      else:
        stringGivenHam=stringGivenHam*(1-value)
    spamGivenString=(stringGivenSpam*pSpam+1)/((stringGivenSpam*pSpam+1)+(stringGivenHam*pHam+1))
    hamGivenString=(stringGivenHam*pHam+1)/((stringGivenSpam*pSpam+1)+(stringGivenHam*pHam+1))
    if(spamGivenString>hamGivenString):
      ans[str1]='spam'
    else:
      ans[str1]='ham'
    if(category.iloc[i]=='spam'):
      actualSpam+=1
      if(category.iloc[i]==ans[str1]):
        preSpam+=1
    if(category.iloc[i]=='ham'):
      actualHam+=1
      if(category.iloc[i]==ans[str1]):
        preHam+=1
    if(category.iloc[i]==ans[str1]):
      count+=1
      
  accuracy=(count/len(testingDataSet))*100
  return accuracy, actualSpam, actualSpam-preSpam, actualHam, preHam
```

```
accuracy, actualSpam, preSpam, actualHam, preHam=testDataAccuracy(spamProb, hamProb, testingDataSet)
print(accuracy)
print('                   ', 'Actual Spam    ','Actual Ham')
print('Predicted Spam     ', preSpam,'        ', actualSpam-preSpam)
print('Predicted Ham      ', actualHam-preHam, '        ',preHam   )
# for key, value in ans.items():
#   if(value=='spam'):
#     print(key)
```
